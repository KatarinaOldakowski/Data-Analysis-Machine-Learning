<b> Data Preprocessing </b>

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df
```

```python
# Converting Categorical Data to Numerical

gender_dict = {'Male': 0, 'Female': 1}
ever_married_dict = {'No': 0, 'Yes': 1}
work_type_dict = {'children': 0, 'Never_worked': 1, 'Govt_job': 2, 'Private': 3, 'Self-employed': 4}
residence_type_dict = {'Rural': 0, 'Urban': 1}
smoking_status_dict = {'Unknown': 0, 'never smoked': 1, 'formerly smoked':2, 'smokes': 3}

df['gender'] = df['gender'].map(gender_dict)
df['ever_married'] = df['ever_married'].map(ever_married_dict)
df['work_type'] = df['work_type'].map(work_type_dict)
df['Residence_type'] = df['Residence_type'].map(residence_type_dict)
df['smoking_status'] = df['smoking_status'].map(smoking_status_dict)
```

```python
df
```

```python
# Checking if data is balanced

xs = df['stroke'].value_counts().index
ys = df['stroke'].value_counts().values

ax = sns.barplot(x=xs, y=ys)
ax.set_xlabel("Stroke")
plt.show()
```

```python
# Splitting into features and predictions

X = df.drop(['stroke'],axis=1)
y = df['stroke']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

print('Number transations x_train df',X_train.shape)
print('Number transations x_test df',X_test.shape)
print('Number transations y_train df',y_train.shape)
print('Number transations y_test df',y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE

print('Before OverSampling, counts of label 1: {}'.format(sum(y==1)))
print('Before OverSampling, counts of label 0: {} \n'.format(sum(y==0)))
```

```python
sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_resample(X_train,y_train.ravel())

print('After OverSampling, the shape of train_x: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {}'.format(y_train_res.shape))

print('After OverSampling, counts of label 1: {}'.format(sum(y_train_res == 1)))
print('After OverSampling, counts of label 0: {}'.format(sum(y_train_res == 0)))
```

```python
labels = ['0', '1']
counts = [sum(y_train_res == 1), sum(y_train_res == 0)]


sns.barplot(x=labels, y=counts)
plt.xlabel('Label')
plt.ylabel('Counts')
plt.title('Counts of Labels after OverSampling')
plt.show()
```

```python
# Creating Random Forest Classifier model

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

pipeline = make_pipeline(StandardScaler(), RandomForestClassifier())
pipeline.fit(X_train, y_train)
prediction = pipeline.predict(X_test)

print(f"Accuracy Score : {round(accuracy_score(y_test, prediction) * 100, 2)}%")
```

```python
print(classification_report(y_test, prediction))
```

```python
confusion_log=confusion_matrix(y_test, prediction)
plt.figure(figsize=(8,8))
sns.heatmap(confusion_log,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
print(classification_report(y_test, prediction))
```
